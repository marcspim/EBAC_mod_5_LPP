```python
# Função para realizar as operações matemáticas
def operacao(opcao, num1, num2):
    if opcao == 1:
        return num1 + num2
    elif opcao == 2:
        return num1 - num2
    elif opcao == 3:
        return num1 * num2
    elif opcao == 4:
        while num2 == 0: # Laço que garante que a divisão não seja feita por 0
            print("Divisão por zero não permitida!")
            num2 = float(input("Digite outro número: "))
        return num1 / num2
    else:
        return "Opção inválida!"

# Laço que permite ao usuário escolher uma operação repetidamente
while True:
    # Solicita ao usuário inserir dois números dentro do loop
    try:
        num1 = float(input("Digite o primeiro número: "))
        num2 = float(input("Digite o segundo número: "))
    except ValueError:
        print("Por favor, insira um número válido.")
        continue  # Pula a iteração e pede novamente os números

    print("\nEscolha uma operação matemática:")
    print("1. Soma")
    print("2. Subtração")
    print("3. Multiplicação")
    print("4. Divisão")
    print("5. Sair")

    try:
        opcao = int(input("Digite o número da operação (1-5): "))
    except ValueError:
        print("Opção inválida! Por favor, digite um número entre 1 e 5.")
        continue  # Pula a iteração e pede a opção novamente

    # Caso o usuário escolha a opção 5, o programa é encerrado
    if opcao == 5:
        print("Saindo...")
        break

    # Realiza a operação e exibe o resultado
    resultado = operacao(opcao, num1, num2)
    print(f"O resultado da operação é: {resultado}.")
```
